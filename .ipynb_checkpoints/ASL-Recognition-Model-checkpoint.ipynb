```python
from pathlib import Path

# Data Processing
import numpy as np
import pandas as pd

# Image Processing
import mediapipe as mp
import imageio as iio
import cv2

# ML libraries
from sklearn import preprocessing
import tensorflow as tf
import sklearn.metrics as skm

# Plotting libraries
import matplotlib.pyplot as plt
import seaborn as sn
```

# Preprocessing

```python
# Function to get extract hand via mediapipe
def getHandCoordinates(image):
    
    handsModule = mp.solutions.hands
    drawingModule = mp.solutions.drawing_utils
    
    with handsModule.Hands(static_image_mode=True) as hands:
        results = hands.process(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
        h,w,c = image.shape
        x_landmarks = []
        y_landmarks = []
        
        # In case no hand is detected
        if results.multi_hand_landmarks == None:
            return -1,-1,-1,-1
        
        # In case hand is detected
        for handLandmarks in results.multi_hand_landmarks:
            for landmark in handLandmarks.landmark:
                x_landmarks.append(landmark.x)
                y_landmarks.append(landmark.y)
                
    # Add 5% padding to the hand
    min_x = int(min(x_landmarks)*w*0.95)
    min_y = int(min(y_landmarks)*h*0.95)
    max_x = int(max(x_landmarks)*w*1.05)
    max_y = int(max(y_landmarks)*h*1.05)
    
    # returns the coordinates of the hand + padding
    return min_x,min_y,max_x,max_y
```

```python
#load images
image_path = Path("./dataset")
images = []
sample_size = 1 # Take every fifth image

# Iterate over data folders
for filename in image_path.iterdir():
    
    video_num = 0
    currentLetter = filename.name
    print(currentLetter)
    video_file = Path(f"{filename}/videos")
    
    # Iterate over videos
    for video in video_file.iterdir():
        # Get frames
        frames = iio.get_reader(video)
        frame_count = -1 
        
        # Iterate over frames
        for frame in frames:
            frame_count += 1
            if frame_count % sample_size != 0:
                continue
            # Convert to RGB color space
            frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
            # Get hand coordinates
            min_x,min_y,max_x,max_y = getHandCoordinates(frame)
            # If no hand is detected, continue to next frame
            if min_x == -1:
                continue
            # Crop hand out
            hand_frame = frame[min_y:max_y,min_x:max_x]
            try: # Compress frame to save space
                hand_frame = cv2.resize(hand_frame,dsize=(32,32))
            except:
                break
            images.append([hand_frame, currentLetter, video_num])
        video_num += 1
```

```python
# Split dataset
images_df = pd.DataFrame(images, columns=["frame", "gesture", "video_id"])
train_img = images_df[images_df["video_id"] <= 24].sample(frac=1)
test_img = images_df[images_df["video_id"] > 24].sample(frac=1)
```

```python
# Labelencode the gestures
le = preprocessing.LabelEncoder()
le.fit(train_img['gesture'])
train_img['gesture'] = le.transform(train_img['gesture'])
test_img['gesture'] = le.transform(test_img['gesture'])
```

```python
# Save Labelencoder
#np.save('labelencoder.npy', le.classes_)
```

```python
# Convert to Tensors
X_train = tf.convert_to_tensor(train_img["frame"].tolist())
X_test =  tf.convert_to_tensor(test_img["frame"].tolist())
y_train = tf.convert_to_tensor(train_img["gesture"].tolist())
y_test = tf.convert_to_tensor(test_img["gesture"].tolist())
```

# Convolutional Neural Network

```python
# Building a CNN to convert to images to an output letter
model = tf.keras.Sequential()

# Data Augmentation
model.add(tf.keras.layers.experimental.preprocessing.RandomFlip("horizontal"))
model.add(tf.keras.layers.experimental.preprocessing.RandomRotation((-0.05,0.05)))
model.add(tf.keras.layers.Conv2D(6, kernel_size=(3,3), activation='relu', input_shape=(32,32,3)))
model.add(tf.keras.layers.AveragePooling2D())
model.add(tf.keras.layers.Conv2D(16, kernel_size=(3,3), activation='relu'))
model.add(tf.keras.layers.AveragePooling2D())
model.add(tf.keras.layers.Dropout(0.2))

# flatten
model.add(tf.keras.layers.Flatten())

# hidden layers
model.add(tf.keras.layers.Dense(120, activation='relu'))
model.add(tf.keras.layers.Dense(84, activation='relu'))

# output layer
model.add(tf.keras.layers.Dense(6, activation='softmax'))

# compiling the sequential model
model.compile(loss=tf.keras.losses.sparse_categorical_crossentropy, metrics=[tf.keras.metrics.SparseCategoricalAccuracy()], optimizer="adam")

history = model.fit(X_train, y_train, epochs=50, validation_data=(X_test,y_test))
```

```python
model.evaluate(X_test,y_test)
```

```python
# Save model
# model.save("./model")
```

# Evaluation

```python
# Plot History
# summarize history for accuracy
plt.plot(history.history['sparse_categorical_accuracy'])
plt.plot(history.history['val_sparse_categorical_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()

# summarize history for loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
# Plot Confustion Matrix
y_test_rounded_pred = tf.argmax(model.predict(X_test), axis = 1)

labels =["A","B","C","L","R","U"]
x = skm.confusion_matrix(y_true=y_test,
                        y_pred=y_test_rounded_pred,
                        normalize="true")
fig, ax = plt.subplots(figsize=(8,6))   
df = pd.DataFrame(x,labels,columns=labels)
sn.heatmap(df, annot=True,ax=ax, fmt='g')
plt.show()
```
